```python
from bs4 import BeautifulSoup
import requests
import os 
import re
import urllib
import csv
```

```python
#資料class: list-unstyled  圖片col-md-5 col-lg-6
def main():
    file = open('store_data.csv',mode='w', newline='',encoding="UTF-8")
    writer = csv.writer(file)
    writer.writerow(["畫作資訊","圖片連結"])
    for result in range(1,2001):
        resp = requests.get('https://www.ss.net.tw/paint-169-'+str(result)+'.html')
        soup = BeautifulSoup(resp.text, 'html.parser')
        T1=soup.find(class_='mb-3')
        B3 = soup.find(class_='list-unstyled')
        print(B3.getText())       
        imgList = soup.find('img','lazyload')
        print(imgList.get('data-src'))
        writer.writerow([B3.getText(),imgList.get('data-src')])
file.close()
```

```python
main()
```
